```python
# Download image
!gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

import matplotlib . image as mpimg
import numpy as np

img = mpimg.imread ('/content/dog.jpeg')

def lightness_method(img):
  return (np.max(img, axis=2) + np.min(img, axis=2)) / 2
   # axis = 2 là dành cho kênh màu RGB

gray_img_01 = lightness_method(img)
gray_img_01 [0 , 0]
# A
```
